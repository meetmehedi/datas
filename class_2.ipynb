```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, Normalizer # Importing the scaler
from sklearn.impute import SimpleImputer # Importing the imputer

# Load your data (replace 'cardio.csv' with your actual file path)
data = pd.read_csv('cardio.csv')

# Impute missing values using the mean
# Create an imputer instance
imputer = SimpleImputer(strategy='mean')

# Fit and transform the imputer on your data
data_imputed = imputer.fit_transform(data)

# Convert the imputed data back to a Pandas DataFrame
data_imputed = pd.DataFrame(data_imputed, columns=data.columns)


# Apply Normalizer to the imputed data
scaler = Normalizer()
data_s= scaler.fit_transform(data_imputed) # Applying Normalizer to the imputed data
print(data_imputed) # Printing the imputed data
sc_data = pd.DataFrame(data_s,columns=data.columns)
sc_data.head()

# Define x and y using the scaled data
x = sc_data.drop('cardio', axis=1)
y = sc_data['cardio']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```
